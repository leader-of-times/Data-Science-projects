# **ZOMATO DATA ANALYSIS PROJECT**


**Importing all necessary libraries with alias names**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**Creating Data Frame and printing it**

```python
dataframe = pd.read_csv("Zomato data .csv")
print(dataframe)
```

```python
dataframe
```

**Converting the data type of column - rate**

```python
def handleRate(value):
  value = str(value).split('/')
  value = value[0]
  return float(value)

dataframe['rate'] = dataframe['rate'].apply(handleRate)
print(dataframe.head())
```

# **Exploratory Data analysis (EDA)**

1. What type of restaurants do the majority of customers order from?

2. How many votes has each type of restaurant recieved from customers?

3. What are the ratings that the majority of restaurants have received?

4. Zomato has observed that most couples order most of their food online. What is the average spending on each order?

5. Which mode (online or offline) has recieved the maximum rating?

6. Which type of restaurant recieved more offline orders, so that Zomato can pay customers with some good offers?

```python
dataframe.head()
```

```python
print("Question 1 :\n ")
sns.countplot(x=dataframe['listed_in(type)'])
plt.xlabel('Type of restaurant')
```

**Conclusion - Majority of the restaurant falls under the dining category**

```python
dataframe.head()
```

```python
print("Question 2: \n")
grouped_data = dataframe.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes':grouped_data})
plt.plot(result,c = "blue", marker = "o")
plt.xlabel("Type of restaurant",c = "black",size = 20)
plt.ylabel("Votes", c = "black", size = 20)
```

**Conclusion - Dining restaurants has recieved maximum votes**

```python
dataframe.head()
```

```python
print("Question 3: \n")
plt.hist(dataframe['rate'],bins = 50)
plt.title("Ratings Distribution")
plt.show()
```

**Conclusion - The majority restaurants recieved ratings from 3.5 to 4**

```python
dataframe.head()
```

```python
print("Question 4: \n")
couple_data = dataframe['approx_cost(for two people)']
sns.countplot(x = couple_data)
```

**Conclusion - The majority of couples prefer restaurants with an approximate cost of â‚¹300**

```python
dataframe.head()
```

```python
print("Question 5: \n")
plt.figure(figsize = (6,6))
sns.boxplot(x ='online_order', y = 'rate',data = dataframe)
```

**Conclusion - Online orders has recieved maximum ratings than offline**

```python
dataframe.head()
```

```python
print("Question 6: \n")
pivot_table = dataframe.pivot_table(index = 'listed_in(type)', columns = 'online_order', aggfunc = 'size', fill_value = 0)
sns.heatmap(pivot_table, annot = True , cmap = "YlGnBu",fmt = 'd')
plt.title("Heatmap")
plt.xlabel("Online Order")
plt.ylabel("Listed In (Type)")
plt.show()
```

**Conclusion - Dining restaurants primarily accept offline orders, wheread cafes primarily recieve online orders.**
